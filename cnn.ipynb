## Свёрточные нейронные сети

### Архитектура свёрточной нейросети

- Нейроны разбиты на слои.
- Когда нейронная сеть применяется к данным, значения этих слоёв (активация данных) подсчитывается последовательно: сначала значения активации первого слоя, потом значения активации второго слоя итд. 
- Есть параметры, которые определяют, как активация следующего слоя зависит от активации предыдущего слоя. 
- Активации внутри разных слоёв могут подсчитываться параллельно, они друг от друга не зависят.

### Обучение нейросети

Настройка параметров (обучение нейросети) происходит следующим образом:
* Берем текущее состояние нейросети, применяем ее к конкретному обучающему примеру и смотрим, что получается на выходе и как результат отличается от того, что мы хотели бы увидеть.
* По методу обратного распространения ошибки можно понять, *как именно* модифицировать веса, чтобы на выходе было что-то более похожее на то, что мы хотим увидеть.

<img src="images/ex9.png" alt="Drawing" style="width: 500px;"/>

Особенность свёрточной нейросети в том, что на первых слоях нейроны разбиты на изображения определенного размера, называемые *картами*.
Разные карты внутри одного слоя соответствует нейронам разного типа, которые реагируют на разные особенности изображений. 
Вычисление активаций следующего слоя бывают двух основных типов:

- Активация нейронов вычисляется как **линейная комбинация** нейронов предыдущего уровня;
- Активация нейронов следующего уровня **повторяет активацию нейронов предыдщего уровня** засчет того, что активация рядом расположенных нейронов заменяется на их максимум или на их среднее.

## U-NET

#### Зачем

Чтобы обучать нейросети на малом количестве изображений высокой четкости.

#### Что делает

Принимает на вход изображение и возвращает его разбитое на сегменты изображение (возаращает карту сегментации).       
Другими словами: **каждый пиксель ставится в соответствии одному из заданных классов.**

![title](images/u-net.png)

**Архитектура сети**

Состоит из **сжимающего пути** (вдоль которого изображение уменьшается) и **расширяющего пути** (вдоль которого изображение увеличивается). 

*Сжимающий путь* (слева) - типичен для свёрточных нейросетей. Он состоит из двух свёрток размером 3x3, каждая из которых сопровождается функцией активации ReLU и двумя пулингами для понижения размерности. При каждом понижении размерности мы удваиваем число признаков каналов изображения.  
На выходе сжимающего пути получается сжатое отображение картинки.

Каждый шаг в *расширяющем пути* предискретизируется. То есть данные увеличиваются в размерах засчёт интерполяции:
<img src="images/ex11.jpeg" alt="Drawing" style="width: 300px;"/>

Или обыденного увеличения размера пикселей (*Pixel Shuffle*), когда при помощи свёртки изображение размерами X*Y*3 заменяется изображением X*Y*12, и пикселей становится в 4 раза больше:
<img src="images/ex12.jpeg" alt="Drawing" style="width: 500px;"/>

То, что приходит с расширяющего пути на каждой итерации "склеивается" с тем, что приходит с соответствующей итерации сжимающего пути (на это указывает чёрная стрелочка-пунктир на изображении архитектуры).
